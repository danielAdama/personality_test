**Goal:** Is to group the personality of Users which is an unsupervised learning Task.

- EXT - Extraversion (outgoing/energitic)
- ARG - Agreeableness (friendly/compassionate)
- CSN - Conscientiousness (efficient/organized)
- OPN - Openness to experience (inventive/curious)
- EST - Neuroticism (sensitive/nervous)

```python
samples = samples.drop(['country'], axis = 1)
cols = samples.columns
feats = samples[cols]
# Normalize and fit samples
# scaler = StandardScaler().fit(feats.values)
# feats = scaler.transform(feats.values)
feats = Normalizer().fit_transform(feats.values)
scaled_feats = pd.DataFrame(feats, columns = cols)

df_sample = scaled_feats[:5000]
kmeans = KMeans()
visualizer = KElbowVisualizer(kmeans, k = (2, 15))
visualizer.fit(df_sample)
visualizer.poof()
plt.show()
```

As you can see 5 cluster looks optimal for the dataset which is the value of k = 5.

## Cluster Participants into 5 personality groups

```python
df_analyze = scaled_feats.copy()
# Define 5 clusters and fit the Model
model = KMeans(n_clusters = 5, random_state = 42)
model.fit(df_analyze)
```

```python
np.unique(model.labels_)
```

```python
# Predicting the Clusters
predictions = model.labels_
df_analyze['Clusters'] = predictions
df_analyze.head()
```

```python
df_analyze.Clusters.value_counts()
```

# Evaluate the Clusters

```python
df_analyze.groupby('Clusters').mean()
```

```python
# Let's sum up each question group(EXT, EST..etc) to check for a pattern.
# Summing up the different questions groups
col_list = list(df_analyze)
ext = col_list[0:10]
est = col_list[10:20]
agr = col_list[20:30]
csn = col_list[30:40]
opn = col_list[40:50]

data_sums = pd.DataFrame()
data_sums['extroversion'] = df_analyze[ext].sum(axis=1)/10
data_sums['neurotic'] = df_analyze[est].sum(axis=1)/10
data_sums['agreeable'] = df_analyze[agr].sum(axis=1)/10
data_sums['conscientious'] = df_analyze[csn].sum(axis=1)/10
data_sums['open'] = df_analyze[opn].sum(axis=1)/10
data_sums['clusters'] = predictions
data_sums.groupby('clusters').mean()
```

```python
# Visualizing the means for each cluster
dataclusters = data_sums.groupby('clusters').mean()
plt.figure(figsize=(22,3))
for i in range(0, 5):
    plt.subplot(1,5,i+1)
    plt.bar(dataclusters.columns, dataclusters.iloc[:, i], color='green', alpha=0.2)
    plt.plot(dataclusters.columns, dataclusters.iloc[:, i], color='red')
    plt.title('Cluster ' + str(i))
    plt.xticks(rotation=45);
```

```python
# We will use PCA to visualize in 2D graph.
df_analyze = samples.copy()
pca = PCA(n_components = 2, random_state = 42)
pca_fit = pca.fit_transform(df_analyze)
df_pca = pd.DataFrame(data = pca_fit, columns = ['PCA1', 'PCA2'])
df_pca['Clusters'] = predictions
df_pca.head()
```

```python
plt.figure(figsize = (10, 10))
sns.set(style = 'white')
sns.scatterplot(data = df_pca, x = 'PCA1', y = 'PCA2', hue = 'Clusters', 
                palette = 'Set2', alpha = 0.8)
plt.title('Personality Cluster after PCA')
plt.show()
```
